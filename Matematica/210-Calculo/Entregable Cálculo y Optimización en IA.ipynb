```python
# Ex  1
import math
import numpy as np

import sympy as sp
from sympy.utilities.lambdify import lambdify # Import lambdify
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = sp.symbols('x')

g1 = x**3 + 3*x**2 -2
g2 = x**2*sp.sin(x) + x
g3 = 4*x**2 - 10*x + 4
g4 = x**4 - 4*x**3 + 4

g1_prime = sp.diff(g1, x)
g2_prime = sp.diff(g2, x)
g3_prime = sp.diff(g3, x)
g4_prime = sp.diff(g4, x)

def gradient_descent(f, df, initial_x, learning_rate, iterations):
    
    x = initial_x
    y =f(x)
    trajectory = [x]
    
    for i in range(iterations):
        x = x - learning_rate * df(x)
        y = f(x)
        trajectory.append(x)
    
    return x ,y, trajectory

# g1
g1_numerical = lambdify(x, g1, 'numpy')
dg1_numerical = lambdify(x, g1_prime, 'numpy')

x0 = 10
learning_rate = 0.01
iterations = 10

x_min_g1, g1_xmin_val, gd_g1_trajectory = gradient_descent(g1_numerical, dg1_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de g1:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_g1}")
print(f"  Final g1(x_min): {g1_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_g1 = np.array(gd_g1_trajectory)
trajectory_y_g1 = g1_numerical(trajectory_x_g1)

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

# Plot g1, g1_prime, and the gradient descent trajectory using matplotlib
print("Plotting g1, g1_prime, and gradient descent trajectory for g1...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, g1_numerical(x_vals_plot), label='g1(x)', color='blue')
plt.plot(x_vals_plot, dg1_numerical(x_vals_plot), label='g1_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_g1, trajectory_y_g1, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('g1, g1_prime, and GD Trajectory (g1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

#g2 

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

g2_numerical = lambdify(x, g2, 'numpy')
dg2_numerical = lambdify(x, g2_prime, 'numpy')

x0 = -5
learning_rate = 0.01
iterations = 100
x_min_g2, g2_xmin_val, gd_g2_trajectory = gradient_descent(g2_numerical, dg2_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de g2:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_g2}")
print(f"  Final g2(x_min): {g2_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_g2 = np.array(gd_g2_trajectory)
trajectory_y_g2 = g2_numerical(trajectory_x_g2)

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

# Plot g2, g2_prime, and the gradient descent trajectory using matplotlib
print("Plotting g2, g2_prime, and gradient descent trajectory for g2...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, g2_numerical(x_vals_plot), label='g2(x)', color='blue')
plt.plot(x_vals_plot, dg2_numerical(x_vals_plot), label='g2_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_g2, trajectory_y_g2, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('g2, g2_prime, and GD Trajectory (g2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# g3

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

g3_numerical = lambdify(x, g3, 'numpy')
dg3_numerical = lambdify(x, g3_prime, 'numpy')

x0 = 8
learning_rate = 0.01
iterations = 100
x_min_g3, g3_xmin_val, gd_g3_trajectory = gradient_descent(g3_numerical, dg3_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de g3:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_g3}")
print(f"  Final g3(x_min): {g3_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_g3 = np.array(gd_g3_trajectory)
trajectory_y_g3 = g3_numerical(trajectory_x_g3)

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

# Plot g3, g3_prime, and the gradient descent trajectory using matplotlib
print("Plotting g3, g3_prime, and gradient descent trajectory for g3...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, g3_numerical(x_vals_plot), label='g3(x)', color='blue')
plt.plot(x_vals_plot, dg3_numerical(x_vals_plot), label='g3_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_g3, trajectory_y_g3, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('g3, g3_prime, and GD Trajectory (g3)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# g4

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

g4_numerical = lambdify(x, g4, 'numpy')
dg4_numerical = lambdify(x, g4_prime, 'numpy')

x0 = -10
learning_rate = 0.001
iterations = 100
x_min_g4, g4_xmin_val, gd_g4_trajectory = gradient_descent(g4_numerical, dg4_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de g4:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_g4}")
print(f"  Final g4(x_min): {g4_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_g4 = np.array(gd_g4_trajectory)
trajectory_y_g4 = g4_numerical(trajectory_x_g4)

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

# Plot g4, g4_prime, and the gradient descent trajectory using matplotlib
print("Plotting g4, g4_prime, and gradient descent trajectory for g4...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, g4_numerical(x_vals_plot), label='g4(x)', color='blue')
plt.plot(x_vals_plot, dg4_numerical(x_vals_plot), label='g4_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_g4, trajectory_y_g4, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('g4, g4_prime, and GD Trajectory (g4)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Ex 2
import math
import numpy as np
import sympy as sp
from sympy.utilities.lambdify import lambdify # Import lambdify
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = sp.symbols('x')

h1 = sp.cos(x) + x**2
h2 = sp.exp(-x)*(x**2+5*x)
h3 = sp.ln(x+1) + x**2
h4 = x**5 -5*x**3+4*x

# h primes

h1_prime = sp.diff(h1, x)
print("La derivada de h1(x) ser: ", h1_prime)

h2_prime = sp.diff(h2, x)
print("La derivada de h2(x) ser: ", h2_prime)

h3_prime = sp.diff(h3, x)
print("La derivada de h3(x) ser: ", h3_prime)

h4_prime = sp.diff(h4, x)
print("La derivada de h4(x) ser: ", h4_prime)


# Gradient descent

def gradient_descent(f, df, initial_x, learning_rate, iterations):
    
    x = initial_x
    y =f(x)
    trajectory = [x]
    
    for i in range(iterations):
        x = x - learning_rate * df(x)
        y = f(x)
        trajectory.append(x)
    
    return x ,y, trajectory

# GD Application h1

h1_numerical = lambdify(x, h1, 'numpy')
dh1_numerical = lambdify(x, h1_prime, 'numpy')

x0 = 10
learning_rate = 0.05
iterations = 100

x_min_h1, h1_xmin_val, gd_h1_trajectory = gradient_descent(h1_numerical, dh1_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de h1:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_h1}")
print(f"  Final h1(x_min): {h1_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_h1 = np.array(gd_h1_trajectory)
trajectory_y_h1 = h1_numerical(trajectory_x_h1)

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

# Plot g1, g1_prime, and the gradient descent trajectory using matplotlib
print("Plotting h1, h1_prime, and gradient descent trajectory for h1...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, h1_numerical(x_vals_plot), label='h1(x)', color='blue')
plt.plot(x_vals_plot, dh1_numerical(x_vals_plot), label='h1_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_h1, trajectory_y_h1, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('h1, h1_prime, and GD Trajectory (h1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

#h2

h2_numerical = lambdify(x, h2, 'numpy')
dh2_numerical = lambdify(x, h2_prime, 'numpy')

x0 = 1
learning_rate = 0.01
iterations = 1000

x_min_h2, h2_xmin_val, gd_h2_trajectory = gradient_descent(h2_numerical, dh2_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de h2:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_h2}")
print(f"  Final h2(x_min): {h2_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_h2 = np.array(gd_h2_trajectory)
trajectory_y_h2 = h2_numerical(trajectory_x_h2)

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 400)

# Plot g1, g1_prime, and the gradient descent trajectory using matplotlib
print("Plotting h2, h2_prime, and gradient descent trajectory for h2...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, h2_numerical(x_vals_plot), label='h2(x)', color='blue')
plt.plot(x_vals_plot, dh2_numerical(x_vals_plot), label='h2_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_h2, trajectory_y_h2, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('h2, h2_prime, and GD Trajectory (h2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

#h3

h3_numerical = lambdify(x, h3, 'numpy')
dh3_numerical = lambdify(x, h3_prime, 'numpy')

x0 = 1
learning_rate = 0.01
iterations = 100

x_min_h3, h3_xmin_val, gd_h3_trajectory = gradient_descent(h3_numerical, dh3_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de h3:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_h3}")
print(f"  Final h3(x_min): {h3_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_h3 = np.array(gd_h3_trajectory)
trajectory_y_h3 = h3_numerical(trajectory_x_h3)

# Generate x values for plotting
x_vals_plot = np.linspace(-10, 10, 100)

# Plot g1, g1_prime, and the gradient descent trajectory using matplotlib
print("Plotting h3, h3_prime, and gradient descent trajectory for h3...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, h3_numerical(x_vals_plot), label='h3(x)', color='blue')
plt.plot(x_vals_plot, dh3_numerical(x_vals_plot), label='h3_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_h3, trajectory_y_h3, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('h3, h3_prime, and GD Trajectory (h3)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

#h3

h4_numerical = lambdify(x, h4, 'numpy')
dh4_numerical = lambdify(x, h4_prime, 'numpy')

x0 = 1
learning_rate = 0.1
iterations = 15000

x_min_h4, h4_xmin_val, gd_h4_trajectory = gradient_descent(h4_numerical, dh4_numerical, x0, learning_rate, iterations)

print(f"\nGradient Descent de h4:")
print(f"  Initial x: {x0}")
print(f"  Learning Rate: {learning_rate}")
print(f"  Iterations: {iterations}")
print(f"  Final x_min: {x_min_h4}")
print(f"  Final h4(x_min): {h4_xmin_val}")

# Prepare trajectory for plotting: get x and y coordinates
trajectory_x_h4 = np.array(gd_h4_trajectory)
trajectory_y_h4 = h4_numerical(trajectory_x_h4)

# Generate x values for plotting
x_vals_plot = np.linspace(-5, 5, 10)

# Plot g1, g1_prime, and the gradient descent trajectory using matplotlib
print("Plotting h4, h4_prime, and gradient descent trajectory for h4...")
plt.figure(figsize=(8, 6))
plt.plot(x_vals_plot, h4_numerical(x_vals_plot), label='h4(x)', color='blue')
plt.plot(x_vals_plot, dh4_numerical(x_vals_plot), label='h4_prime(x)', color='green', linestyle='--')
plt.plot(trajectory_x_h4, trajectory_y_h4, label='GD Trajectory', color='red', marker='o', linestyle='-', markersize=4)
plt.title('h4, h4_prime, and GD Trajectory (h4)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
